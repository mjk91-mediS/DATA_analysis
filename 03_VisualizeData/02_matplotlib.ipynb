# 실습 문제

```python
import matplotlib.pyplot as plt
import pandas as pd

stock_df = pd.read_csv("../00_data/stock_daily_prices.csv")
stock_df["Date"] = pd.to_datetime(stock_df["Date"])
print(stock_df.info())
```

# 문제 1: 
테슬라(TSLA) 주가의 시간에 따른 변동을 시각화하고, 이를 애플(AAPL) 주가와 비교하는 선 그래프를 그리세요.

```python
plt.figure(figsize=(10, 6))
#plot(): 선 그래프를 그리는 함수이다. 
# stock_df['Date']: X축에 날짜(Date)를 설정하고, y축에 Stock_df[Tsla]를 표시한다.
plt.plot(stock_df['Date'], stock_df['TSLA'], label='TSLA', color='r')
plt.plot(stock_df['Date'], stock_df['AAPL'], label='AAPL', color='b')

plt.title('TSLA vs AAPL Stock Price Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

# 문제 2: 
아마존(AMZN) 주가의 분포를 히스토그램으로 그리세요.


```python
plt.figure(figsize=(8, 6))

# .hist 히스토그램을 그리는 함수이다. 히스토그램은 데이터의 분포를 시각적으로 나타내는 데 사용된다.
# y는 데이터의 분포 x는 데이터의 값값
plt.hist(stock_df['AMZN'], bins=20, color='g', edgecolor='black')

plt.title('AMZN Stock Price Distribution', fontsize=16)
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

# 문제 3: 
2012년 1월 25일의 주식 일일 변동률을 막대 그래프로 시각화

```python
# Date 열을 인덱스로 설정
stock_df = stock_df.set_index("Date")
# 2012-01-25의 데이터 필터링
daily_returns_0125 = stock_df.loc['2012-01-25']

# 차트 크기 설정
plt.figure(figsize=(10, 6))

# 주식별 변동률을 다른 색상으로 막대 그래프 생성
# kind = "bar" 그래프의 종류를 의미하는 속성으로 bar 형식으로 만든다.
# color = plt.cm.viridis.colors 색상을 의미하는 속성으로 plt.cm.viridis.colors 색상 팔레트이다.
daily_returns_0125.plot(kind='bar', color="purple", edgecolor='black',legend=False)

# 제목 및 축 레이블 설정
plt.title('Stock Returns on 2012-01-25', fontsize=16)
plt.xlabel('Stock', fontsize=12)
plt.ylabel('Daily Return (%)', fontsize=12)

# 그리드 추가
plt.grid(True)

# 그래프 출력
plt.show()
```
